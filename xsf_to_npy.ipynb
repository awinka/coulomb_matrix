Script for creating npy-files from xsf-files. The npy-files are faster for 
reading and therefore benefitial.

```python
import glob
import numpy as np
from xsf_2_np import Xsf2Np

#folder = "/usr/scratch/bucaramanga/awinka/MoS2/principle_unit_cell/wannier/no_gw_opt_no_symmetry/"
#save_folder = "/usr/scratch/bucaramanga/awinka/MoS2/principle_unit_cell/wannier/no_gw_opt_no_symmetry/np_files/"
#folder = "/usr/scratch/bucaramanga/awinka/MoS2/wannier/orth/wannier/"
#save_folder = "/usr/scratch/bucaramanga/awinka/MoS2/wannier/orth/wannier/npy_files/"
#folder = "/usr/scratch/bucaramanga/awinka/MoS2/principle_unit_cell/wannier/no_gw_opt_more_vb_with_real_space/"
#save_folder = "/usr/scratch/bucaramanga/awinka/MoS2/principle_unit_cell/wannier/no_gw_opt_more_vb_with_real_space/npy_files/"
folder = "/usr/scratch/bucaramanga/awinka/MoS2/principle_unit_cell/wannier/no_gw_strain5/"
save_folder = "/usr/scratch/bucaramanga/awinka/MoS2/principle_unit_cell/wannier/no_gw_strain5/npy_files/"
files = glob.glob(folder + "/*.xsf")
for file in files:
    print(f"Processing file: {file}")
    with open(file, 'r') as f:
        xsf2np = Xsf2Np(f)
    np.save(save_folder + file.split("/")[-1].split(".")[0] + ".npy", xsf2np.WF)
```
