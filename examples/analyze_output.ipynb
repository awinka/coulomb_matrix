```python
import numpy as np
import matplotlib.pyplot as plt

coulomb_matrix = np.load("outputs/coulomb_matrix.npy")
exchange_matrix = np.load("outputs/exchange_matrix.npy")
```

```python
wf1, wf2 = 0, 0  # Example wavefunction indices to plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(coulomb_matrix[0,0,:, wf1, wf2], label=f'Coulomb, WF {wf1} vs {wf2}')
ax.plot(exchange_matrix[0,0,:, wf1, wf2], label=f'Exchange, WF {wf1} vs {wf2}', linestyle='--')
ax.set_xlabel('Index')
ax.set_ylabel('Matrix Value')
ax.set_title('Coulomb and Exchange Matrix Comparison')
ax.legend()
plt.show()
```

```python
# Print the minimum and maximum values of the matrices for comparison
print(f"Coulomb Matrix: min={np.min(coulomb_matrix)} eV, max={np.max(coulomb_matrix)} eV")
print(f"Exchange Matrix: min={np.min(exchange_matrix)} eV, max={np.max(exchange_matrix)} eV")
```
